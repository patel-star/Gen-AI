```python
mask = np.zeros_like(image)

height = image.shape[0]
width = image.shape[1]

center = (width // 2, height // 2)
rad = min(center[0], center[1])

cv2.circle(mask, center, rad, (255,255,255), -1)

masked_image = np.bitwise_and(image, mask)
plt.imshow(masked_image)
```

```python
# Intensity Histogram

image = cv2.imread(image_paths[0])
greyscale_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(greyscale_image)

plt.hist(greyscale_image.ravel(), 256, [0,256])
plt.show()
```

```python
image = cv2.imread(image_paths[0])
color = ('b', 'g', 'r')

for i, col in enumerate(color):
    histr = cv2.calcHist([image], [i], None, [256], [0,256])
    plt.plot(histr, color = col)
    plt.xlim([0,256])

plt.title('Color Histogram')
plt.show()
```

```python
image_paths = ['/content/drive/MyDrive/GenAI - 2025/DataSets/cat_dog_lite/training_set/dogs/dog.17.jpg']
```

```python
img = cv2.imread(image_paths[0])
# plt.imshow(img)
img
```

```python
blurred = cv2.blur(img, (5,5))
plt.imshow(blurred)
```

```python
cv2.Canny(blurred, 40, 120)
```

```python
translatex = 100
translatey = 50

moved_image = cv2.warpAffine(img, np.float32([[1,0,translatex], [0,1,translatey]]), (img.shape[1], img.shape[0]))
plt.imshow(moved_image)
```

```python
angle = 45
scale = 1

ht,wd = img.shape[:2]
rotation_matrix = cv2.getRotationMatrix2D((wd/2, ht/2), angle, scale)

rotated_image = cv2.warpAffine(img, rotation_matrix, (wd, ht))
plt.imshow(rotated_image)
```
